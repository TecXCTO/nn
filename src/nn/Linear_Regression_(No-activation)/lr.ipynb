```python
import numpy as np

# Simple linear regression: y = wx + b
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

w = np.random.randn(1)
b = 0.0
lr = 0.01 #
print("w:", w, "b:", b, "lr:",lr)

for epoch in range(200):
    y_pred = X @ w + b 
    # loss = ((y_pred - y) ** 2).mean() #
    # dw = (2/len(X)) * (X.T @ (y_pred - y))
    # db = (2/len(X)) * (y_pred - y).sum()
    print("epoch =", epoch)
    print("X =",X)
    print("y =",y)
    print("y_pred =",y_pred) #
    y_err= y_pred - y #
    print("y_err =",y_err) #
    y_err_sqr = y_err ** 2 #
    print("y_err_sqr =",y_err_sqr) #
    loss = (y_err_sqr).mean() #
    print("loss =", loss) #
    print("len(X) =",len(X)) #
    X_T = X.T #
    print("X_T=", X_T) #
    x_T_dot_yerr = X_T@y_err #
    print("x_T@yerr =", x_T_dot_yerr) #
    dw = (2/len(X)) * x_T_dot_yerr #
    print("dw =",dw) #
    y_err_sum = (y_err).sum() #
    print("y_err_sum", y_err_sum) #
    db = (2/len(X)) * y_err_sum #
    print("db =",db) #
    w -= lr * dw
    b -= lr * db

print("w:", w, "b:", b)
```
