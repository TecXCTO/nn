```python
import numpy as np

# Simple linear regression: y = wÂ·x + b
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

w = np.random.randn(1)
b = 0.0
lr = 0.01

for epoch in range(200):
    y_pred = X @ w + b
    loss = ((y_pred - y) ** 2).mean()
    dw = (2/len(X)) * (X.T @ (y_pred - y))
    db = (2/len(X)) * (y_pred - y).sum()
    w -= lr * dw
    b -= lr * db

print("w:", w, "b:", b)
```
